# list in data structures 
1. list is collection of elements enclosed by [ ]
2. list a mutable - editable
3. it's allowed  duplicate  elements
4. each element seperated by ,

```python
# create list
names=["shiva","vikram","vikas","divya"]
ages=[12,15,21,27.5,29.8]
print(names)
print(ages)
print(type(names))
print(type(ages))
```

```python
#concatination of list
a=[10,11,12,13,14]
b=[15,16,17,18,19]
result=a+b
print(result)
```

```python
#repeating list
ages=[10,20,30,40,50]
print(ages*15)
```

```python
# accessing elements from list - index
ages=[15,16,17,18,19,25,26]
print(ages)
print(ages[4])
print(ages[-3])
```

```python
#methods or function in list
print(dir(list))
```

```python
#discription
print(help(list))
```

```python
#add elements to list-append(),extend(),insert()
a=[10,20,30,40,50]
print(a)
a.append(100)
print(a)
```

```python
a=[10,20,30,40,50]
print(a)
a.extend([100,200,300])
print(a)
```

```python
a=[10,20,30,40,50]
print(a)
a.insert(1,100)
print(a)
```

```python
#remove elements from the list - pop(),remove(),clear()
a=[10,20,30,40,50]
print(a)
a.pop(-1)
print(a)
```

```python
a=[10,20,30,40,50]
print(a)
a.remove(50)
print(a)
```

```python
# sorting techniques -sort(reverse=True or reverse=Talse)
a=[6,7,8,9,10]
print(a)
a.sort(reverse=True) # True means decending order, false means accending order
print(a)
```

# tuples in data structures
1. tuple is collection of elements enclosed by ( )
3. tuple is mutable -  not editable
4. it's allowed  duplicate  elements
5. each element seperated by ,
   

```python
a=(10,20,30,40,50)
names=("shiva","vikram",10.25,12.56)
print(a)
print(names)
print(type(a))
print(type(names))
```

```python
#concatenation,repeating,accesing elements from tuple
a=(1,2,3)
b=(4,5,6,7)
combined= a+b
print(combined)
c=(8,9)
extended=combined+c
print(extended)
```

```python
#repeating
a=("hello, ")
repeat=a*5
print(repeat)
```

```python
# accessing elements
a=(1,2,3,4,5,6,7,8,9)
print(a[2:])
print(a[1])
print(a[0:9])
```

```python
#methods
print(dir(tuple))
```

```python
# count ()
a=(10,10,20,30,10,20,40,30)
print(a.count(10))
print(a.count(20))
print(a.count(30))
```

```python
#index()
a=(10,20,30,40,50)
print(a.index(40))
```
