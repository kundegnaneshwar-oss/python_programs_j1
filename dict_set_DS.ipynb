# dict
1. dict is collection of items enclosed by {} - item is a key:value pair

```python
info={"name":"shiva","salary":100000.0,"address":"hyderabad","age":25}
print(info)
print(type(info))
```

```python
# read  data and updated data from dict
info={"name":"shiva","salary":100000.0,"address":"hyderabad","age":25}
print(info)
print(info["address"])
```

```python
# update
info={"name":"shiva","salary":100000.0,"address":"hyderabad","age":25}
print(info)
info["address"]="banglore"
print(info)
```

```python
# methods or function in dict
print(dir(dict))
```

```python
info={"name":"shiva","salary":100000.0,"address":"hyderabad","age":25}
print(info)
print(info.keys())
print(info.values())
print(info.items())
```

```python
# update()
info={"name":"shiva","salary":100000.0,"address":"hyderabad","age":25}
info.update({"salary":2000000})
print(info)
```

```python
# pop()
info={"name":"shiva","salary":100000.0,"address":"hyderabad","age":25}
print(info)
info.pop("salary")
print(info)
```

# set
1. set is collection of unorder unique elements enclosed by{}

```python
# create set
a={2,4,1,3,5,6,7,1,2,3}
print(a)
print(type(a))
```

```python
# methods or functions
print(dir(set))
```

```python
# add()
a={20,30,10,40}
print(a)
a.add(100)
print(a)
```

```python
#difference()
a={1,2,3,4,5,6}
b={4,5,6,7,8,9}
c=a.difference(b)
print(c)
d=b.difference(a)
print(d)
```

```python
#union()
a={1,2,3,4,5,6}
b={4,5,6,7,8,9}
c=a.union(b)
print(c)
```

```python
#intersection()
a={1,2,3,4,5,6}
b={4,5,6,7,8,9}
c=a.intersection(b)
print(c)
```

```python
#discard
a={1,2,3,4,5,6}
print(a)
a.discard(6)
print(a)
```
