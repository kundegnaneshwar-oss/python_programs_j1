# conditional statement in python
## python have 4 statements of conditional statement
1. `if` statement
2. `elif` statement
3. `else` statement
4. `nested` statement

```python - if
if (condition): # if condition true inside statements are running
    statement1
    statement...
```

```python -if,else
if (condition):#if condition true inside statements are running
    statement1
    statements2...n
else: #if condition fails its went to else statement and print inside statements
```

````python
# write a program to find the given number is even number?
number=int(input("enter number here:"))
if (number%2 == 0):
    print(number,"is even number")
````

````python
# write a program to find the given number is even number or odd number?
num = int(input("please enter number here"))
if (num%2 == 1):
    print(num,"is odd number")
else:
    print(num,"is even number")
````

````python
# write a program to find the given number is positive or negative or zero?
n = int(input("please enter number"))
if (n > 0):
    print(n,"is positive number")
elif(n < 0):
    print(n,"is negative number")
else:
    print(n,"is zero")
````

````python
#write a program to find the largest number from three numbers?
v1= int(input("enter first number here:"))
v2= int(input("enter second  here:"))
v3= int(input("enter third number here:"))
if(v1>=v2 and v1>=v3):
    print(v1," is largest number")
elif(v2>=v1 and v2>=v3):
    print(v2," is largest number")
else:
    print(v3," is largest number")
````
