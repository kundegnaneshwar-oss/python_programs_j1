# operators in python

## python have 6 types of operators
1. arithematic operators
2. assignment operators
3. comparision operators
4. logical operators
5. membership operatos
6. identity operators

```python
#arithematic operators - +,-,*,/,//,%
a=10
b=20
c=30
d=40
print(a+b)
print(b-d)
print(a*c)
print(b/a)
print(b//a)
print(d%b)
```

```python
#assignment operators - +=,-=,*=,/=,//=,%=
a=10
a+=10
print(a)
```

```python
a=25
a-=50
print(a)
```

```python
a=50
a*=100
print(a)
```

```python
a=40
a/=20
print(a)
```

```python
a=80
a//=40
print(a)
```

```python
a=200
a%=100
print(a)
```

```python
#comarision operators - >,<,>=,<=,!= (return ture or false)
a=5
b=10
c=10
d=20
print(a>b)
print(b<c)
print(b>=c)
print(b==c)
print(b!=d)
```

```python
#logical operators - and,or,not
# and - this operator checking all conditions true result is true rest cases failedd
# or - this operator check either one conditions true  true then result true rest cases failed
a=10
b=20
c=20
d=60
print(a>b and b==c)
print(a<b and c<d)
print(a>b or b==c)
print(a>b or b!=c)
print(not a>b or b!=c)
```

```python
#membership operators - checkin members are present in that sequence or collection - in,not in
names=["gnaneshwar","vikas","madhav","divya"]
print("divya" in names)
print("alok" in names)
print("Gnaneshwar" not in names)
print("Vikas" in names)
```

```python
# identity operators - memory identification of values -is,is not
a=10000
b=10000
print(id(a))
print(id(b))
print(a is b)
```

```python
a=10
b=10
print(id(a))
print(id(b))
print(a is b)
```
