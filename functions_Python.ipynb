````python
# write a program to print sum of two numbers using functions
def add_number():
    a=int(input("enter first number"))
    b=int(input("enter second number"))
    result=a+b
    print(result)
    return result
add_number()
````

````python
# write a progarm to print addition of two strings using function
def add_string():
    a=str(input("enter surname"))
    b=str(input("enter  first name"))
    result=a+b
    print(result)
    return result
add_string()
````
