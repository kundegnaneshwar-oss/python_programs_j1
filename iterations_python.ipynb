```python
#write a program to print sum of all even numbers and odd numbers using for loop and while loop?
se=0
so=0
for n in range(1,101):
    if(n%2==0):
        se=se+n
    elif(n%2==1):
        so=so+n
print("sum of even numbers in between 1 to 100 is",se)
print("sum of odd numbers in between 1 to 100 is",so)    
```
